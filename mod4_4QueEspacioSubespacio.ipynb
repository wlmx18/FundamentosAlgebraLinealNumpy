```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])

for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b,v1[1]*a + v2[1]*b, marker='.',color='orange')
plt.xlim(-25,25)
plt.ylim(-25,25)
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
        
```

```python
#Obtenemos una recta ya que definimos las coordenadas como 1, R2 fue una ilusion
#Si nos abstraemos que estamos en R2 y solamente hubieramos entregado a x para graficar hubieramos obtenido lo mismo
#Dimos elementos para R2 pero obtenimos R1
#R1 es una parte de R2
```
