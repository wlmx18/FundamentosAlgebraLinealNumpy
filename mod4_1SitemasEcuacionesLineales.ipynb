```python
#configurar matplotlib de manera interactiva
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definir el intervalo donde vamos estar evaluando cada unos de los valores que toma y 
x = np.arange(-6,6)

y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1
```

```python
#SIN SOLUCION
#generamos la figura donde vamos a estar mostrando nuestro grafico
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
#limites de los ejes
plt.xlim(-8,8)
plt.ylim(-8,8)

#configuracion de la posicion y color del eje
plt.axvline(x=0,color='grey')
plt.axhline(y=0,color='grey')

#muestrame los graficos
plt.show()
```

```python
#UNA SOLUCION
x = np.arange(-6,6)

y_2 = -x + 3
y_3 = 2*x + 1

#generamos la figura donde vamos a estar mostrando nuestro grafico
plt.figure()

plt.plot(x, y_2)
plt.plot(x, y_3)
#limites de los ejes
plt.xlim(-8,8)
plt.ylim(-8,8)

#configuracion de la posicion y color del eje
plt.axvline(x=0,color='grey')
plt.axhline(y=0,color='grey')

#muestrame los graficos
plt.show()
```
